### Introducción
Este estudio analiza la transición de la matriz eléctrica colombiana, históricamente dependiente de la generación hídrica (~70%). Se evalúa cómo la integración de fuentes no convencionales (FNCER) y la geotermia actúan como mecanismos de resiliencia frente a la variabilidad climática (Fenómeno de El Niño).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Simulación de carga de datos (Aquí conectarás los archivos que descargaste de XM)
# Imagina que ya uniste los archivos de las carpetas 2020 a 2024
data = {
    'Año': [2020, 2021, 2022, 2023, 2024, 2025, 2026],
    'Hidrica_GWh': [120, 118, 122, 115, 110, 105, 100], # Tendencia a la baja por diversificación
    'Solar_GWh': [1, 2, 5, 12, 25, 45, 60],            # Crecimiento exponencial
    'Eolica_GWh': [0.5, 0.5, 1, 3, 10, 25, 40],        # Reto: Entrada de proyectos en Guajira
    'Geotermica_GWh': [0, 0, 0, 0, 0.5, 2, 5]          # Oportunidad: Pilotos Ecopetrol/Nereidas
}

df = pd.DataFrame(data)

# 2. Cálculo de Participación (Diversificación)
df['Total_Generacion'] = df.sum(axis=1) - df['Año']
df['%_Renovables_No_Conv'] = ((df['Solar_GWh'] + df['Eolica_GWh'] + df['Geotermica_GWh']) / df['Total_Generacion']) * 100

# 3. Visualización de la Transición 2020-2026
plt.figure(figsize=(12,6))
plt.stackplot(df['Año'], df['Solar_GWh'], df['Eolica_GWh'], df['Geotermica_GWh'],
              labels=['Solar', 'Eólica', 'Geotérmica'], colors=['#FFD700', '#87CEEB', '#FF4500'], alpha=0.8)
plt.title('Crecimiento de Energías Alternativas en Colombia (Proyectado 2026)')
plt.ylabel('Generación (GWh)')
plt.legend(loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

print(f"En 2020 la participación de renovables era del {df['%_Renovables_No_Conv'].iloc[0]:.2f}%")
print(f"Para 2026 se proyecta una participación del {df['%_Renovables_No_Conv'].iloc[-1]:.2f}%")
```

**Reasoning**:
Following the instructions, the next step is to initialize an empty list to store the generated data and then iterate 1000 times, generating synthetic data for each record based on the specified parameters, trends, and constraints. This involves selecting a random year and energy type, calculating base metrics using linear interpolation between defined start and end values for each energy type, adding random variations, and ensuring logical consistency of the generated values before appending them to the list.



```python
data_records = []

# Define trends for each energy type and metric
# (Start Year: 2020, End Year: 2026)

trends = {
    'Solar': {
        'Generacion_GWh': {'start': 1, 'end': 60, 'min_var': 0.8, 'max_var': 1.2},
        'Oferta_GWh': {'start': 1.2, 'end': 65, 'min_var': 0.9, 'max_var': 1.1},
        'Demanda_GWh': {'start': 0.8, 'end': 55, 'min_var': 0.9, 'max_var': 1.0},
        'Costo_MWh': {'start': 50, 'end': 30, 'min_var': 0.9, 'max_var': 1.1},
        'Porcentaje_Cobertura': {'start': 0.01, 'end': 0.15, 'min_var': 0.8, 'max_var': 1.2},
        'Inversion_USD_millones': {'start': 100, 'end': 500, 'min_var': 0.8, 'max_var': 1.2},
        'Emisiones_CO2_toneladas': {'start': 5, 'end': 10, 'min_var': 0.8, 'max_var': 1.2} # Low emissions
    },
    'Eólica': {
        'Generacion_GWh': {'start': 0.5, 'end': 40, 'min_var': 0.8, 'max_var': 1.2},
        'Oferta_GWh': {'start': 0.6, 'end': 45, 'min_var': 0.9, 'max_var': 1.1},
        'Demanda_GWh': {'start': 0.4, 'end': 38, 'min_var': 0.9, 'max_var': 1.0},
        'Costo_MWh': {'start': 60, 'end': 40, 'min_var': 0.9, 'max_var': 1.1},
        'Porcentaje_Cobertura': {'start': 0.005, 'end': 0.1, 'min_var': 0.8, 'max_var': 1.2},
        'Inversion_USD_millones': {'start': 80, 'end': 400, 'min_var': 0.8, 'max_var': 1.2},
        'Emisiones_CO2_toneladas': {'start': 3, 'end': 8, 'min_var': 0.8, 'max_var': 1.2} # Low emissions
    },
    'Geotérmica': {
        'Generacion_GWh': {'start': 0, 'end': 5, 'min_var': 0.8, 'max_var': 1.2},
        'Oferta_GWh': {'start': 0, 'end': 6, 'min_var': 0.9, 'max_var': 1.1},
        'Demanda_GWh': {'start': 0, 'end': 4, 'min_var': 0.9, 'max_var': 1.0},
        'Costo_MWh': {'start': 70, 'end': 50, 'min_var': 0.9, 'max_var': 1.1},
        'Porcentaje_Cobertura': {'start': 0, 'end': 0.02, 'min_var': 0.8, 'max_var': 1.2},
        'Inversion_USD_millones': {'start': 50, 'end': 200, 'min_var': 0.8, 'max_var': 1.2},
        'Emisiones_CO2_toneladas': {'start': 10, 'end': 20, 'min_var': 0.8, 'max_var': 1.2} # Moderate emissions
    },
    'Hídrica': {
        'Generacion_GWh': {'start': 120, 'end': 100, 'min_var': 0.8, 'max_var': 1.2},
        'Oferta_GWh': {'start': 125, 'end': 105, 'min_var': 0.9, 'max_var': 1.1},
        'Demanda_GWh': {'start': 110, 'end': 95, 'min_var': 0.9, 'max_var': 1.0},
        'Costo_MWh': {'start': 20, 'end': 25, 'min_var': 0.9, 'max_var': 1.1},
        'Porcentaje_Cobertura': {'start': 0.7, 'end': 0.5, 'min_var': 0.8, 'max_var': 1.2},
        'Inversion_USD_millones': {'start': 10, 'end': 50, 'min_var': 0.8, 'max_var': 1.2},
        'Emisiones_CO2_toneladas': {'start': 50, 'end': 60, 'min_var': 0.8, 'max_var': 1.2} # Higher emissions (from construction, maintenance, etc.)
    }
}

# Interpolation function
def linear_interpolate(start_val, end_val, current_year, start_year=2020, end_year=2026):
    if start_year == end_year:
        return start_val
    progress = (current_year - start_year) / (end_year - start_year)
    return start_val + progress * (end_val - start_val)

for _ in range(num_records):
    year = random.choice(years)
    energy_type = random.choice(energy_types)

    record = {'Año': year, 'Tipo_Energia': energy_type}

    for metric, props in trends[energy_type].items():
        base_value = linear_interpolate(props['start'], props['end'], year)

        # Add random variation
        value = base_value * random.uniform(props['min_var'], props['max_var'])

        # Ensure non-negative values for relevant metrics
        if metric not in ['Costo_MWh', 'Porcentaje_Cobertura']:
            value = max(0, value)

        record[metric] = value

    # Apply logical constraints
    record['Oferta_GWh'] = max(record['Oferta_GWh'], record['Generacion_GWh'] * 1.05) # Oferta >= Generacion * 1.05
    record['Demanda_GWh'] = min(record['Demanda_GWh'], record['Oferta_GWh'] * random.uniform(0.8, 0.95)) # Demanda <= Oferta
    record['Porcentaje_Cobertura'] = np.clip(record['Porcentaje_Cobertura'], 0.0, 1.0) # Cobertura between 0 and 1
    record['Costo_MWh'] = max(0, record['Costo_MWh']) # Costo no negativo
    record['Inversion_USD_millones'] = max(0, record['Inversion_USD_millones']) # Inversion no negativa
    record['Emisiones_CO2_toneladas'] = max(0, record['Emisiones_CO2_toneladas']) # Emisiones no negativas

    data_records.append(record)

df_synthetic = pd.DataFrame(data_records)

print("Generated synthetic DataFrame:")
print(df_synthetic.head())
print("\nDataFrame Info:")
df_synthetic.info()
```

```python
df_comparativo = df_synthetic.groupby(['Año', 'Tipo_Energia']).agg({
    'Generacion_GWh': 'mean',
    'Oferta_GWh': 'mean',
    'Demanda_GWh': 'mean',
    'Costo_MWh': 'mean',
    'Porcentaje_Cobertura': 'mean',
    'Inversion_USD_millones': 'mean',
    'Emisiones_CO2_toneladas': 'mean'
}).reset_index()

print("DataFrame comparativo (primeras filas):")
print(df_comparativo.head())

print("\nInformación del DataFrame comparativo:")
df_comparativo.info()
```
